```python
#adding resultants and calculated rotational velocities to new data file
#IMPORTANT:delete 'FULL' files before re-running
import os
import glob
from pathlib import Path
import pandas as pd

#path = 'D:/MASTERS/K_files_2020'
path = 'D:\MASTERS\K_files_2020_duplicates_removed'

filelist = os.listdir(path)
print(filelist)
#loop through all files.
#reading the file

for value in filelist:
    kinematics = pd.read_csv(path + '/' + value)
    #print(path + '/' + value)
    #print(kinematics)

    #creating empty array for rotational velocity
    rv_x = []
    rv_y = []
    rv_z = []

    for k in range(len(kinematics)):
        #calculating resultant rotational acceleration
        if k == 0:
            rv_x.append(0)
            rv_y.append(0)
            rv_z.append(0)
        else:
            rv_x_temp = ((kinematics['ra_x'][k-1] + kinematics['ra_x'][k])*0.00004*1000)/2 + rv_x[k-1] # trapezium rule h(a+b)*1/2 (*1000 to convert untis) + last velocity
            rv_y_temp = ((kinematics['ra_y'][k-1] + kinematics['ra_y'][k])*0.00004*1000)/2 + rv_y[k-1]
            rv_z_temp = ((kinematics['ra_z'][k-1] + kinematics['ra_z'][k])*0.00004*1000)/2 + rv_z[k-1]
            #appending to the arrays
            rv_x.append(rv_x_temp)
            rv_y.append(rv_y_temp)
            rv_z.append(rv_z_temp)

    kinematics['rv_x'] = rv_x
    kinematics['rv_y'] = rv_y
    kinematics['rv_z'] = rv_z
    
    #creating empty list for resultants
    ra_r = []
    la_r = []
    rv_r = []

    for k in range(len(kinematics)):
        #calculating resultant rotational acceleration
        ra_resultant_temp = ((kinematics['ra_x'][k])**2 + (kinematics['ra_y'][k])**2 + (kinematics['ra_z'][k])**2) ** (0.5)
        ra_r.append(ra_resultant_temp)
        #calculating resultant linear acceleration
        la_resultant_temp = ((kinematics['la_x'][k])**2 + (kinematics['la_y'][k])**2 + (kinematics['la_z'][k])**2) ** (0.5)
        la_r.append(la_resultant_temp)
        #calculating resultant rotational velocity
        rv_resultant_temp = ((kinematics['rv_x'][k])**2 + (kinematics['rv_y'][k])**2 + (kinematics['rv_z'][k])**2) ** (0.5)
        rv_r.append(rv_resultant_temp)
    #add to kinematics 
    kinematics['ra_r'] = ra_r
    kinematics['la_r'] = la_r
    kinematics['rv_r'] = rv_r

    #print(kinematics)

    df = pd.DataFrame(kinematics)
    df.to_csv(path + '/' + 'FULL_' + value, index=False)
    print(path + '/' + 'FULL_' + value )

```
