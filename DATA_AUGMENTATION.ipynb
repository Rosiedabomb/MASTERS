```python
import pandas as pd
import numpy as np
path = "E:/MASTERS/IMPACT_DATA.csv"

```

```python
#read data
raw_data = pd.read_csv(path)
#flip X,X rot accelerations & velocities flip Y linear velocities.
flips = ['peak y lin acc','peak x rot acc', 'peak x rot vel','peak z rot acc','peak z rot vel']
augmented_data = raw_data.copy(deep=True) #copy will not affect original
for value in flips:
    augmented_data[value] = -1* raw_data[value]
#print(augmented_data)
#print(raw_data)

#add 'augment' label to impact label
print(len(augmented_data['Impact label']))
for v in range(len(augmented_data['Impact label'])):
    augmented_data['Impact label'][v] = augmented_data['Impact label'][v] + '_augmented'
#add datasets together somehow.
combined_data = pd.concat([raw_data, augmented_data], ignore_index=True)
print(combined_data)
```

```python
#save to new csv.
df = pd.DataFrame(augmented_data)
df.to_csv('E:/MASTERS/IMPACT_DATA_AUGMENTED_ONLY.csv', index=False)
print('augmented data csv saved')

DF = pd.DataFrame(combined_data)
DF.to_csv('E:/MASTERS/IMPACT_DATA_AUGMENTED.csv', index=False)
print('combined data csv created')
```
